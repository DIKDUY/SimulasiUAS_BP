```python
class RawMaterialLogistics:
    def __init__(self):
        self.orders = {}
        self.order_id_counter = 1

    def create_order(self, supplier_name, material_name, quantity):
        order_id = self.order_id_counter
        self.orders[order_id] = {
            'supplier_name': supplier_name,
            'material_name': material_name,
            'quantity': quantity,
            'status': 'Pending'
        }
        self.order_id_counter += 1
        return order_id

    def update_status(self, order_id, status):
        if order_id in self.orders:
            self.orders[order_id]['status'] = status
            return True
        return False

    def get_order(self, order_id):
        return self.orders.get(order_id, None)


# Inisialisasi layanan logistik
logistics_service = RawMaterialLogistics()

# Fungsi untuk menampilkan menu dan interaksi
def logistics_menu():
    while True:
        print("\n=== Jasa Penggadaan Bahan Baku ===")
        print("1. Create Order")
        print("2. Update Order Status")
        print("3. Get Order Information")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            supplier_name = input("Enter supplier name: ")
            material_name = input("Enter material name: ")
            quantity = int(input("Enter quantity: "))
            order_id = logistics_service.create_order(supplier_name, material_name, quantity)
            print(f"Order created successfully with ID: {order_id}")

        elif choice == '2':
            order_id = int(input("Enter order ID: "))
            status = input("Enter new status: ")
            if logistics_service.update_status(order_id, status):
                print("Order status updated successfully")
            else:
                print("Order not found")

        elif choice == '3':
            order_id = int(input("Enter order ID: "))
            order = logistics_service.get_order(order_id)
            if order:
                print(f"Order Information: ID={order_id}, Supplier={order['supplier_name']}, Material={order['material_name']}, Quantity={order['quantity']}, Status={order['status']}")
            else:
                print("Order not found")

        elif choice == '4':
            break

        else:
            print("Invalid choice. Please try again.")
```
